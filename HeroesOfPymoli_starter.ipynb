```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
grouped_items = purchase_data.groupby(["Item ID"])
average_price = purchase_data["Price"].mean()

unique_items =  pd.DataFrame({
    "Item Name": grouped_items["Item Name"].first(),
    "Times Purchased": grouped_items["Item ID"].count(),
    "Price": grouped_items["Price"].first(),
    "Average Age of Purchaser": grouped_items["Age"].mean(),
})

unique_items.Price = unique_items.Price.map("${:.2f}".format)
unique_items["Average Age of Purchaser"] = unique_items["Average Age of Purchaser"].map("{:.1f}".format)
game_data = pd.DataFrame([
    {
        "unique_items":unique_items
    }
    ]
)

unique_items.head()
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
def toPct(count, total):
    return f"{round(count / total * 100,2)}%"

#get unique users
grouped_by_user = purchase_data.groupby(["SN"])

unique_users =  pd.DataFrame({
    "User Name": grouped_by_user["SN"].first(),
    "Gender": grouped_by_user["Gender"].first(),
    "Age": grouped_by_user["Age"].first(),
})

gender_counts = unique_users["Gender"].value_counts()

gender_demographics = {
    "Male": toPct(gender_counts["Male"],player_count),
    "Female": toPct(gender_counts["Female"],player_count),
    "Other / Non-Disclosed": toPct(gender_counts["Other / Non-Disclosed"],player_count),
}
gender_demographics
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
grouped_by_gender = purchase_data.groupby(["Gender"])

gender_analysis =  pd.DataFrame({
    "Purchase Count": grouped_by_gender["Gender"].count(),
    "Average Price": grouped_by_gender["Price"].mean().map("${:.2f}".format),
    "Average Purchase Total": round(grouped_by_gender["Gender"].count() / gender_counts,2)
})
#print(gender_counts)
gender_analysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
max_age = purchase_data["Age"].max()
age_bins = [0,17,24,34,49,10000]
bin_names = ["Under 18","18 to 24","25 to 34","34 to 49","50+"]

unique_users["Age Demographic"] = pd.cut(unique_users["Age"],age_bins,labels=bin_names)

group_by_demo = unique_users.groupby(["Age Demographic"])

age_demo_df = pd.DataFrame({
    "Count": group_by_demo["Age"].count(),
    "Percentage": group_by_demo["Age"].count()/player_count
})
age_demo_df["Percentage"] = ["{:.2f}%".format(val * 100) for val in age_demo_df["Percentage"]]
age_demo_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
all_items = pd.DataFrame({
    "Item ID": purchase_data["Item ID"],
    "Item Name":  purchase_data["Item Name"],
    "Price":  purchase_data["Price"],
})
grouped_items = all_items.groupby(["Item ID", "Item Name"])

item_raw_data = pd.DataFrame({
    "Purchase Count":grouped_items["Price"].count(),
    "Price": grouped_items["Price"].first(),
    "Total Purchase Value": grouped_items["Price"].sum(),
})
item_raw_data = item_raw_data.sort_values(by=['Purchase Count'], ascending=False)
#item_popularity_analysis = item_raw_data

item_popularity_analysis = pd.DataFrame()

item_popularity_analysis["Purchase Count"] = item_raw_data["Purchase Count"]
item_popularity_analysis["Price"] = item_raw_data["Price"].map("${:.2f}".format)
item_popularity_analysis["Total Purchase Value"] = item_raw_data["Total Purchase Value"].map("${:.2f}".format)

item_popularity_analysis.head(20)
```
